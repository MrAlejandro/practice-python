```python
from pandas_datareader import data
from bokeh.plotting import figure, show, output_file
import datetime

def get_record_status(open_value, close_value):
    if open_value > close_value:
        status = 'Decrease'
    elif open_value < close_value:
        status = 'Increase'
    else:
        status = 'Equal'
        
    return status

start = datetime.datetime(2016,6,1)
end = datetime.datetime(2017,6,10)

data_frame = data.DataReader(name='GOOG', data_source='google', start=start, end=end)
data_frame['Status'] = [get_record_status(o, c) for o, c in zip(data_frame['Open'], data_frame['Close'])]
data_frame['Height'] = [abs(o - c) for o, c in zip(data_frame['Open'], data_frame['Close'])]
data_frame['Mid'] = [abs(o + c)/2 for o, c in zip(data_frame['Open'], data_frame['Close'])]
data_frame
```

```python
page = figure(x_axis_type='datetime', width=1000, height=500, title='Candlestick Chart', responsive=True)
twelve_hours_in_msec = 12 * 60 * 60 * 1000

page.segment(data_frame.index, data_frame.High, data_frame.index, data_frame.Low, color='black')

page.rect(
    data_frame.index[data_frame.Status == 'Increase'],
    data_frame.Mid[data_frame.Status == 'Increase'],
    twelve_hours_in_msec,
    data_frame.Height[data_frame.Status == 'Increase'],
    fill_color='green',
    line_color='black'
)

page.rect(
    data_frame.index[data_frame.Status == 'Decrease'],
    data_frame.Mid[data_frame.Status == 'Decrease'],
    twelve_hours_in_msec,
    data_frame.Height[data_frame.Status == 'Decrease'],
    fill_color='red',
    line_color='black'
)

output_file('CS.html')
show(page)
```
