```python
import requests
from bs4 import BeautifulSoup
```

```python
request = requests.get('http://pythonhow.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/')
parsed = BeautifulSoup(request.content, 'html.parser')
```

```python
property_rows = parsed.find_all('div', {'class':'propertyRow'})
property_rows[0].find('h4', {'class':'propPrice'}).text.strip(' \t\n\r')
```

```python
property_list = []

for row in property_rows:
    row_data = {}
    
    row_data['price'] = row.find('h4', {'class':'propPrice'}).text.strip(' \t\n\r')
    row_data['address'] = (', '.join(
            [
                address.text.strip(' \t\n\r') 
                for address 
                in row.find_all('span', {'class':'propAddressCollapse'})
            ]
        )
    )
    
    sec_details = row.find('div', {'class':'secondaryDetails'})
    
    try:
        row_data['beds_number'] = sec_details.find('span', {'class': 'infoBed'}).find('b').text
    except:
        row_data['beds_number'] = None
    
    try:
        row_data['area'] = sec_details.find('span', {'class': 'infoSqFt'}).find('b').text
    except:
        row_data['area'] = '';
    
    try:
        row_data['full_baths'] = sec_details.find('span', {'class': 'infoValueFullBath'}).find('b').text
    except:
        row_data['full_baths'] = None
    
    try:
        row_data['half_baths'] = sec_details.find('span', {'class': 'infoValueHalfBath'}).find('b').text
    except:
        row_data['half_baths'] = None
    
    for column_group in row.find_all('div', {'class':'columnGroup'}):
        feature_group = column_group.find('span', {'class':'featureGroup'})
        
        if feature_group and feature_group.text.find('Lot Size') != -1:
            row_data['lot_size'] = column_group.find('span', {'class':'featureName'}).text
            
    if 'lot_size' not in row_data:
        row_data['lot_size'] = None
            
        
    property_list.append(row_data)

len(property_list)
```

```python
import pandas

data_frame = pandas.DataFrame(property_list)
data_frame.to_csv('property.csv')
data_frame
```

```python
import requests, pandas
from bs4 import BeautifulSoup

base_url = 'http://pythonhow.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/t=0&s='
property_list = []

for page in range(0, 30, 10):
    page_url = base_url + str(page) + '.html'
    request = requests.get(page_url)
    parsed = BeautifulSoup(request.content, 'html.parser')
    
    property_rows = parsed.find_all('div', {'class':'propertyRow'})
    
    for row in property_rows:
        row_data = {}

        row_data['price'] = row.find('h4', {'class':'propPrice'}).text.strip(' \t\n\r')
        row_data['address'] = (', '.join(
                [
                    address.text.strip(' \t\n\r') 
                    for address 
                    in row.find_all('span', {'class':'propAddressCollapse'})
                ]
            )
        )

        sec_details = row.find('div', {'class':'secondaryDetails'})

        try:
            row_data['beds_number'] = sec_details.find('span', {'class': 'infoBed'}).find('b').text
        except:
            row_data['beds_number'] = None

        try:
            row_data['area'] = sec_details.find('span', {'class': 'infoSqFt'}).find('b').text
        except:
            row_data['area'] = '';

        try:
            row_data['full_baths'] = sec_details.find('span', {'class': 'infoValueFullBath'}).find('b').text
        except:
            row_data['full_baths'] = None

        try:
            row_data['half_baths'] = sec_details.find('span', {'class': 'infoValueHalfBath'}).find('b').text
        except:
            row_data['half_baths'] = None

        for column_group in row.find_all('div', {'class':'columnGroup'}):
            feature_group = column_group.find('span', {'class':'featureGroup'})

            if feature_group and feature_group.text.find('Lot Size') != -1:
                row_data['lot_size'] = column_group.find('span', {'class':'featureName'}).text

        if 'lot_size' not in row_data:
            row_data['lot_size'] = None


        property_list.append(row_data)

data_frame = pandas.DataFrame(property_list)
data_frame.to_csv('property.csv')
```
